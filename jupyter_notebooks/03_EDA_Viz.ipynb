### Section 4 -Exploratory DAta Analysis (Relationships and insights)
In this section, I explore how key sales-related variables interact with one another and contribute to overal business performance. 
Because this dataset version includes only one numeric metric ('SAles'), correlation analysis is limited -but meaningful insights can be drawn from categorical comparisons. 
The following visualisations focus on:
-  Sales by category:** to identify which product groups drive the most revenue. 
-  Sales by Region** to compare performance across geographical areas. 
- Sales by Segment:** to understand which customer groups contribute most to sales. 
These visualisations help uncover business patterns and potential focus areas for future sales strategy. 

```python
import matplotlib.pyplot as plt 
import seaborn as sns 

# Correlation heatmap (numeric columns only)
plt.figure(figsize=(5,4))
sns.heatmap(df[['Sales',]].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Only Sales column available')
plt.show()
```

## Observations from relationship analysis 
Bacause this dataset version include only one numeric field ('Sales'), the correlation heatmap shows a single variable. 
This confirms that further corrrelation-based exploration is not possible at this stage. 

However, the categorical breakdown that follow-such as sales by **Category**, **REgion** and **SEgment*8-still provide valuable business insights:
- Certain product categories contribute significantly more to overall sales.Regional performance varies across geaographic Zones. 
- The consumer segment remains the dominant contributor to total sales. 
These insights offer a strong foundation for identifying high -performing areas and focusing on strategic improvements. 

### Sales Analysis by Category, Region and Segment
In this section, I analyse how total sales very across key business dimentions - **Product category**, **Geagraphic region**, and **customer segment**. 
These Visualisations help identify which products, markers, and customer types contribute the most to overal revenue. 
- Sales by Category :**Shows whcih product lines generate the highest revenue. 
- Sales by Region:**Highlights regional performanvce variations. 
- Sales by Segment:** Examines Which customer group to total sales.
This breakdown  provides a clear view of the company's main revenue drivers. 

```python
### Sales by Category 
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Category', y='Sales', estimator=sum)
plt.title('Total Sales by Category')
plt.ylabel('Total Sales')
plt.xlabel('Region')
plt.show()

## Sales by Region 
plt.figure(figsize=(8,5))
sns.barplot(data=df,x='Region', y='Sales', estimator=sum)
plt.title('Total Sales by Region')
plt.ylabel('Total Sales')
plt.xlabel('Region')
plt.show()

# Sales by segment
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Segment', y='Sales', estimator=sum)
plt.title('Total Sales by Segment')
plt.ylabel('Total Sales')
plt.xlabel('Segment')
plt.show()
```

### Observations from relationship analysis 
-**Technology** products generate the highest sales, followed by **Furnitures** and **office supplies**. 
- The **west** region performs the best in terms of sales , while the **central*8 region lags behing. 
- The **consumers** segment contributes the majority of total revenue , indicating strong B2C performance compared to Corpotae and Home Office segments. 
- Sales patterns suggest that product type , region and customer segment all significantly influence total sales outcomes. 
-These insights can guide strategic decisions such as inventory allocation , regional markeking focus and customer targeting.
